```python
# Members
gym_member = "Alex Alliton"
preferred_weight_kg = 20.5
highest_reps = 25
membership_active = True

# Minutes spent on three different workout activities
workout_stats = {
    "Alex": [30, 45, 20],
    "Jamie": [30, 30, 60],
    "Taylor": [30, 45, 20]
}

# Calculate totals and store in a separate dictionary
total = {}
for friend, values in workout_stats.items():
    total_value = sum(values)
    total[f"{friend}_Total"] = total_value

# Print totals dictionary
print(total)

# Add totals into workout_stats as well (so we can use it later for checks)
workout_stats.update(total)

# Activity labels
activities = ["Yoga", "Running", "Weightlifting"]

# 2D list storing workout minutes for slicing
workout_list = []
for friend, values in workout_stats.items():
    if "_Total" not in friend:
        workout_list.append(values)

# Print table
print(f"\n{'People':<10} {activities[0]:<10} {activities[1]:<10} {activities[2]:<15}")
for friend, minutes in workout_stats.items():
    if "_Total" not in friend:
        print(f"{friend:<10} {minutes[0]:<10} {minutes[1]:<10} {minutes[2]:<15}")

# Slice examples
print("\nYoga and Running minutes for all friends:")
for row in workout_list:
    print(row[0:2])

print("\nWeightlifting minutes for the last two friends:")
for row in workout_list[-2:]:
    print(row[2])

# Check if any friend's total workout minutes are >= 120
for key, minutes in workout_stats.items():
    if "_Total" in key:
        if minutes >= 120:
            friend_name = key.replace("_Total", "")
            print(f"\nGreat job staying active, {friend_name}!")

# Ask the user to input a friend's name
friend_input = input("\nEnter a friend's name to check their workout stats: ").strip()

# Check if the friend exists in the dictionary
if friend_input in workout_stats:
    # Print individual activity minutes
    minutes = workout_stats[friend_input]
    print(f"\nWorkout minutes for {friend_input}:")
    print(f"Yoga: {minutes[0]}")
    print(f"Running: {minutes[1]}")
    print(f"Weightlifting: {minutes[2]}")
    
    # Print total minutes (look up in the totals dictionary)
    total_minutes = workout_stats.get(f"{friend_input}_Total", sum(minutes))
    print(f"Total Workout Minutes: {total_minutes}")
else:
    print(f"\nFriend {friend_input} not found in the records.")

    # Extract only the totals from the dictionary
totals_only = {key.replace("_Total", ""): minutes for key, minutes in workout_stats.items() if "_Total" in key}

# Find friend with highest total minutes
max_friend = max(totals_only, key=totals_only.get)
max_minutes = totals_only[max_friend]

# Find friend with lowest total minutes
min_friend = min(totals_only, key=totals_only.get)
min_minutes = totals_only[min_friend]

# Print the results
print(f"\nFriend with the highest total workout minutes: {max_friend} ({max_minutes} minutes)")
print(f"Friend with the lowest total workout minutes: {min_friend} ({min_minutes}) minutes")







```
